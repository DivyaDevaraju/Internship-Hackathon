## 	Happiest and Unhappiest Countries in 2021,2022,2023

```python
import pandas as pd

data_2021 = pd.read_csv(r"C:\Users\DELL\Downloads\archive\world-happiness-report-2021.csv")
data_2022 = pd.read_csv(r"C:\Users\DELL\Downloads\archive\world-happiness-report-2021.csv")
data_2023 = pd.read_csv(r"C:\Users\DELL\Downloads\archive\world-happiness-report-2021.csv")

```

```python
def find_happiest_and_unhappiest(data):
    happiest_countries = data.nlargest(10, 'Ladder score')
    unhappiest_countries = data.nsmallest(10, 'Ladder score')
    return happiest_countries, unhappiest_countries
```

```python
happiest_2021, unhappiest_2021 = find_happiest_and_unhappiest(data_2021)
happiest_2022, unhappiest_2022 = find_happiest_and_unhappiest(data_2022)
happiest_2023, unhappiest_2023 = find_happiest_and_unhappiest(data_2023)
```

```python
print("Happiest Countries in 2021:")
print(happiest_2021[['Country name', 'Ladder score']])
```

```python
print("\nUnhappiest Countries in 2021:")
print(unhappiest_2021[['Country name', 'Ladder score']])
```

```python
print("\nHappiest Countries in 2022:")
print(happiest_2022[['Country name', 'Ladder score']])
```

## 	Relationship Between Happiness and Income

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv(r"C:\Users\DELL\Downloads\archive\world-happiness-report-2021.csv")
```

## 	Ladder Score Distribution by Regional Indicator

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv(r"C:\Users\DELL\Downloads\archive\world-happiness-report-2021.csv")

region_scores = data.groupby('Regional indicator')['Ladder score'].mean().reset_index()
region_scores = region_scores.sort_values(by='Ladder score', ascending=False)


plt.figure(figsize=(12, 6))
plt.bar(region_scores['Regional indicator'], region_scores['Ladder score'])
plt.title('Ladder Score Distribution by Regional Indicator')
plt.xlabel('Regional Indicator')
plt.ylabel('Mean Ladder Score')
plt.xticks(rotation=90) 

plt.tight_layout()
plt.show()
```

## 	Ladder Score Distribution by Countries in Map View

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Regional indicator', y='Ladder score', data=data)
plt.xticks(rotation=90)
plt.xlabel('Regional Indicator')
plt.ylabel('Happiness Score (Ladder score)')
plt.title('Ladder Score Distribution by Regional Indicator')
plt.show()
```

```python
import geopandas as gpd 
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
merged = world.set_index('name').join(data.set_index('Country name'))

# Create a map view of Ladder Scores
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
merged.plot(column='Ladder score', cmap='coolwarm', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)
ax.set_title('Ladder Score by Country')
plt.show()
```

## 	Relationship Between Features

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv(r"C:\Users\DELL\Downloads\archive\world-happiness-report-2021.csv")

numerical_features = ['Ladder score', 'Logged GDP per capita', 'Social support', 'Healthy life expectancy', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']

correlation_matrix = data[numerical_features].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Features')
plt.show()
```
